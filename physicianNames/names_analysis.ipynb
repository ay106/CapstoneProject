```python

# CSV file path
csv_file_path = "names_list.csv"

# Writing to CSV
with open(csv_file_path, mode='w', newline='') as csv_file:
    writer = csv.writer(csv_file)
    
    # Write the header if needed
    # writer.writerow(["Name"])  # Uncomment this line if you want to include a header
    
    # Write names to the CSV file
    for name in namesFilter:
        writer.writerow([name])
```

```python
ids_dictionary = {}
failed_queries = set()
count = 0

for person in namesFilter:
    cursor = "*"
    author_name = person.lower()
    # Keep making queries until the request is successful or the maximum number of queries is reached
    while True:
        query_author = f'https://api.openalex.org/authors?search={author_name}&per_page=100&cursor={cursor}'
        
        try:
            response = requests.get(query_author)


            if response.status_code == 200:
                authors = response.json()["results"]

                for author in authors:
                    name = HumanName(author.pop("display_name"))
    
                    author.pop("display_name_alternatives")
                    author.pop("orcid")
                    author.pop("summary_stats")
                    author.pop("x_concepts")
                    author.pop("works_api_url")
                    
                    author["id"] = author["id"][21:]
                    author["name_first"] = name.first if len(name.first) > 1 else None
                    author["name_middle"] = name.middle if len(name.middle) > 1 else None
                    author["name_last"] = name.last if len(name.last) > 1 else None
                    author["works"] = []
                    
                    author["api"] = "openAlex"
                    
                    if author_name not in ids_dictionary:
                        ids_dictionary[author_name] = {}
                    
                    if author["id"] not in ids_dictionary[author_name]:
                        ids_dictionary[author_name][author["id"]] = {}
                    ids_dictionary[author_name][author["id"]] = author


            if not authors:
                break

            cursor = response.json()["meta"]["next_cursor"]
                
        except Exception as e:
            print(e)
            failed_queries.add((author_name, cursor))
            break
    print(f'Number of Ids for {person}: {len(ids_dictionary[person])}')
    count +=1
    print(f'{count}. {author_name} + done')

print(f"Failed queries: {failed_queries}")
with open("TESTINGIDSS.json", "w") as f:
    json.dump(ids_dictionary, f, indent=4)
# use the names to lookup number of ids in dimensionsAI 
# try to find years with lowest numbers of ids from both databases
```

## Isolating ID from OpenAlex

```python
with open("TESTINGIDSS.json", "r") as f:
    ids_dictionary = json.load(f)

openAlexIds = {}
error_names = []  # List to store names that cause errors

for name in ids_dictionary.keys():
    try:
        content = []
        name_dict = ids_dictionary[name]
        for id_key in name_dict.keys():
            id_nameFirst = name_dict[id_key]["name_first"] 
            id_nameLast = name_dict[id_key]["name_last"] 
            indv_id = name_dict[id_key]["id"]
            id_link = name_dict[id_key]["ids"]["openalex"]
            institutions_info = set()
            for institute in name_dict[id_key]["affiliations"]:
                full_institute = institute["institution"]["display_name"]
                institutions_info.add(full_institute)
            tuple_set = (id_nameFirst, id_nameLast, indv_id, id_link)
            content.append(tuple_set)
        openAlexIds[name] = content
    except Exception as e:
        error_names.append(name)
        print(f"Error processing {name}: {e}")

# Now, openAlexIds contains the successful entries, and error_names contains the names that caused errors.
print("Names that caused errors:", error_names)
```
